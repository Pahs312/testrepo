###  In the bottom code the user can input a number of which the function calculates the square root off and then prints it out

```python
input_number = int(input("What number would you like to calculate the square root of? "))

def square_root(input_number): 
    power = input_number**(1/2) 
    print("The square root of " , input_number , " is " , power)
    
square_root(input_number)
```
